```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

LINEAR RELATIONSHIP

```python
x=np.arange(-5.0,5.0,0.1)
y=2.5*x+5

y_noise=np.random.normal(size=x.size)
y_data=y+y_noise

plt.plot(x,y_data,"bo")
plt.plot(x,y,"r")
plt.ylabel("dependent variable")
plt.xlabel("independet variable")
plt.show()
```

CUBIC

```python
x=np.arange(-5.0,5.0,0.1)
y=x**3+x**2+x+3
y_noise=np.random.normal(size=x.size)
y_data=y+y_noise

plt.plot(x,y_data,"ro")
plt.plot(x,y,"b")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

QUADRATIC

```python
x=np.arange(-5.0,5.0,0.1)
y=4*(x**2)
y_noise=np.random.normal(size=x.size)
y_data=y+y_noise

plt.plot(x,y_data,"bo")
plt.plot(x,y,"r")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

Exponential

```python
X = np.arange(-5.0, 5.0, 0.1)

Y= 2*np.exp(X)

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```

𝑌̂ =1/(1+𝑒𝛽_1(𝑋−𝛽_2))

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y
```

```python
beta_1 = 0.10
beta_2 = 1990.0
x_data=df["Year"].values
y_data=df["Value"].values
#logistic function
Y_pred = sigmoid(x_data, beta_1 , beta_2)

#plot initial prediction against datapoints
plt.plot(x_data, Y_pred*15000000000000.)
plt.plot(x_data, y_data, 'ro')
plt.show()
```

```python
x_data=x_data/max(x_data)
```

```python
y_data=y_data/max(y_data)
```

```python
from scipy.optimize import curve_fit
popt,pcurv=curve_fit(sigmoid,x_data,y_data)
print("The parameters:",popt[0],popt[1])
print(pcurv)
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(x_data, y_data, 'ro', label='data')
plt.plot(x,y, linewidth=3.0, label='fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```
